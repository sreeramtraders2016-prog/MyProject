```python
import pandas as pd

# --------------------------------------------------
# 1. LOAD DATA
# --------------------------------------------------
crm_cust_info = pd.read_csv("/content/cust_info.csv")
erp_cust_info = pd.read_csv("/content/ERP_CUST_AZ12.csv")
erp_cust_loc  = pd.read_csv("/content/ERP_LOC_A101.csv")

# --------------------------------------------------
# 2. CLEAN STRING COLUMNS (TRIM SPACES)
# --------------------------------------------------
def clean_df(df):
    return df.apply(lambda col: col.str.strip() if col.dtype == "object" else col)

crm_cust_info = clean_df(crm_cust_info)
erp_cust_info = clean_df(erp_cust_info)
erp_cust_loc  = clean_df(erp_cust_loc)

# --------------------------------------------------
# 3. STANDARDIZE KEYS (VERY IMPORTANT FOR JOIN)
# --------------------------------------------------
# ERP info: NASAW00011000 → AW00011000
erp_cust_info["cst_key"] = erp_cust_info["CID"].str.replace("NAS", "", regex=False)

# ERP location: AW-00011000 → AW00011000
erp_cust_loc["cst_key"] = erp_cust_loc["CID"].str.replace("-", "", regex=False)

# --------------------------------------------------
# 4. MERGE DATASETS
# --------------------------------------------------
# Merge CRM + ERP Info
df = crm_cust_info.merge(
    erp_cust_info[["cst_key", "BDATE", "GEN"]],
    on="cst_key",
    how="left"
)

# Merge with Location data
df = df.merge(
    erp_cust_loc[["cst_key", "CNTRY"]],
    on="cst_key",
    how="left"
)

# --------------------------------------------------
# 5. FIX DATA TYPES
# --------------------------------------------------
df["cst_create_date"] = pd.to_datetime(df["cst_create_date"])

# Handle invalid dates like 9999-xx-xx
df["BDATE"] = pd.to_datetime(df["BDATE"], errors="coerce", format="mixed")

# --------------------------------------------------
# 6. STANDARDIZE CATEGORICAL COLUMNS
# --------------------------------------------------

# Marital Status: M/S → Married/Single
df["cst_marital_status"] = (
    df["cst_marital_status"]
    .map({"M": "Married", "S": "Single"})
    .fillna("Unknown")
)

# Gender from CRM (M/F → Male/Female)
df["cst_gndr"] = df["cst_gndr"].str.upper().map({
    "M": "Male",
    "F": "Female"
})

# Final Gender (ERP priority, fallback to CRM)
df["gender"] = (
    df["GEN"].str.upper().map({
        "MALE": "Male",
        "FEMALE": "Female"
    })
).fillna(df["cst_gndr"]).fillna("Unknown")

# --------------------------------------------------
# 7. CLEAN COUNTRY VALUES
# --------------------------------------------------
df["CNTRY"] = df["CNTRY"].replace({
    "US": "United States",
    "USA": "United States",
    "DE": "Germany"
})

# Handle blanks and nulls
df["CNTRY"] = df["CNTRY"].replace("", pd.NA).fillna("Unknown")

# --------------------------------------------------
# 8. REMOVE INVALID RECORDS
# --------------------------------------------------
# Drop rows where key columns are missing
df = df[
    df["cst_id"].notna() &
    df["cst_key"].notna()
]

# --------------------------------------------------
# 9. CREATE CUSTOMER NAME
# --------------------------------------------------
df["cust_name"] = (
    df["cst_firstname"].fillna("") + " " + df["cst_lastname"].fillna("")
).str.strip()

# Replace empty names
df["cust_name"] = df["cust_name"].replace("", "Unknown")

# --------------------------------------------------
# 10. SELECT FINAL COLUMNS
# --------------------------------------------------
final_df = df[[
    "cst_id",
    "cst_key",
    "cust_name",
    "cst_marital_status",
    "gender",
    "BDATE",
    "CNTRY",
    "cst_create_date"
]]

# --------------------------------------------------
# 11. REMOVE DUPLICATES
# --------------------------------------------------
final_df = final_df.drop_duplicates(subset="cst_id")

# --------------------------------------------------
# 12. DATA QUALITY CHECKS
# --------------------------------------------------

# Check missing values
print("Null values:\n", final_df.isna().sum())

# Check duplicates
print("\nDuplicate cst_id:", final_df.duplicated(subset="cst_id").sum())

# Count unique customers
print("\nUnique customers:", final_df["cst_id"].nunique())

# Preview data
print("\nSample data:\n", final_df.head())
```

```python
# --------------------------------------------------
# 1. IMPORT SPARK
# --------------------------------------------------
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, trim, when, upper, concat_ws, regexp_replace
from pyspark.sql.types import StringType

spark = SparkSession.builder.appName("CustomerETL").getOrCreate()

# --------------------------------------------------
# 2. LOAD DATA
# --------------------------------------------------
crm_cust_info = spark.read.csv("/content/cust_info.csv", header=True, inferSchema=True)
erp_cust_info = spark.read.csv("/content/ERP_CUST_AZ12.csv", header=True, inferSchema=True)
erp_cust_loc  = spark.read.csv("/content/ERP_LOC_A101.csv", header=True, inferSchema=True)

# --------------------------------------------------
# 3. TRIM STRING COLUMNS
# --------------------------------------------------
def trim_all_columns(df):
    return df.select([
        trim(col(c)).alias(c) if dict(df.dtypes)[c] == "string" else col(c)
        for c in df.columns
    ])

crm_cust_info = trim_all_columns(crm_cust_info)
erp_cust_info = trim_all_columns(erp_cust_info)
erp_cust_loc  = trim_all_columns(erp_cust_loc)

# --------------------------------------------------
# 4. STANDARDIZE KEYS
# --------------------------------------------------
# ERP info: remove 'NAS'
erp_cust_info = erp_cust_info.withColumn(
    "cst_key",
    regexp_replace(col("CID"), "NAS", "")
)

# ERP location: remove '-'
erp_cust_loc = erp_cust_loc.withColumn(
    "cst_key",
    regexp_replace(col("CID"), "-", "")
)

# --------------------------------------------------
# 5. MERGE DATASETS
# --------------------------------------------------
df = crm_cust_info.join(
    erp_cust_info.select("cst_key", "BDATE", "GEN"),
    on="cst_key",
    how="left"
)

df = df.join(
    erp_cust_loc.select("cst_key", "CNTRY"),
    on="cst_key",
    how="left"
)

# --------------------------------------------------
# 6. FIX DATE TYPES
# --------------------------------------------------
df = df.withColumn("cst_create_date", col("cst_create_date").cast("date"))

# Handle invalid dates (Spark will convert invalid to null)
df = df.withColumn("BDATE", col("BDATE").cast("date"))

# --------------------------------------------------
# 7. STANDARDIZE MARITAL STATUS
# --------------------------------------------------
df = df.withColumn(
    "cst_marital_status",
    when(col("cst_marital_status") == "M", "Married")
    .when(col("cst_marital_status") == "S", "Single")
    .otherwise("Unknown")
)

# --------------------------------------------------
# 8. STANDARDIZE GENDER
# --------------------------------------------------
# CRM gender
df = df.withColumn(
    "cst_gndr",
    when(upper(col("cst_gndr")) == "M", "Male")
    .when(upper(col("cst_gndr")) == "F", "Female")
)

# Final gender (ERP preferred)
df = df.withColumn(
    "gender",
    when(upper(col("GEN")) == "MALE", "Male")
    .when(upper(col("GEN")) == "FEMALE", "Female")
    .otherwise(col("cst_gndr"))
)

df = df.withColumn(
    "gender",
    when(col("gender").isNull(), "Unknown").otherwise(col("gender"))
)

# --------------------------------------------------
# 9. CLEAN COUNTRY
# --------------------------------------------------
df = df.withColumn(
    "CNTRY",
    when(col("CNTRY") == "US", "United States")
    .when(col("CNTRY") == "USA", "United States")
    .when(col("CNTRY") == "DE", "Germany")
    .otherwise(col("CNTRY"))
)

df = df.withColumn(
    "CNTRY",
    when((col("CNTRY") == "") | col("CNTRY").isNull(), "Unknown")
    .otherwise(col("CNTRY"))
)

# --------------------------------------------------
# 10. REMOVE INVALID RECORDS
# --------------------------------------------------
df = df.filter(
    col("cst_id").isNotNull() & col("cst_key").isNotNull()
)

# --------------------------------------------------
# 11. CREATE CUSTOMER NAME
# --------------------------------------------------
df = df.withColumn(
    "cust_name",
    concat_ws(" ", col("cst_firstname"), col("cst_lastname"))
)

df = df.withColumn(
    "cust_name",
    when(col("cust_name") == "", "Unknown").otherwise(col("cust_name"))
)

# --------------------------------------------------
# 12. SELECT FINAL COLUMNS
# --------------------------------------------------
final_df = df.select(
    "cst_id",
    "cst_key",
    "cust_name",
    "cst_marital_status",
    "gender",
    "BDATE",
    "CNTRY",
    "cst_create_date"
)

# --------------------------------------------------
# 13. REMOVE DUPLICATES
# --------------------------------------------------
final_df = final_df.dropDuplicates(["cst_id"])

# --------------------------------------------------
# 14. DATA QUALITY CHECKS
# --------------------------------------------------

# Null counts
final_df.select([
    col(c).isNull().cast("int").alias(c)
    for c in final_df.columns
]).groupBy().sum().show()

# Duplicate check
print("Duplicates:", final_df.count() - final_df.dropDuplicates(["cst_id"]).count())

# Total records
print("Total records:", final_df.count())

# Preview
final_df.show(5)
```

```python
# --------------------------------------------------
# 1. IMPORTS & SPARK SESSION
# --------------------------------------------------
from pyspark.sql import SparkSession
from pyspark.sql.functions import (
    col, trim, when, upper, concat_ws,
    regexp_replace, broadcast, to_date
)
from pyspark.sql.types import StringType

spark = SparkSession.builder.appName("CustomerETL_Optimized").getOrCreate()

# --------------------------------------------------
# 2. LOAD DATA (ONLY REQUIRED COLUMNS - PERFORMANCE)
# --------------------------------------------------
crm_cust_info = spark.read.csv(
    "/content/cust_info.csv", header=True, inferSchema=True
).select(
    "cst_id","cst_key","cst_firstname","cst_lastname",
    "cst_marital_status","cst_gndr","cst_create_date"
)

erp_cust_info = spark.read.csv(
    "/content/ERP_CUST_AZ12.csv", header=True, inferSchema=True
).select("CID","BDATE","GEN")

erp_cust_loc = spark.read.csv(
    "/content/ERP_LOC_A101.csv", header=True, inferSchema=True
).select("CID","CNTRY")

# --------------------------------------------------
# 3. TRIM STRING COLUMNS (CLEANING)
# --------------------------------------------------
def trim_all_columns(df):
    return df.select([
        trim(col(c)).alias(c) if isinstance(df.schema[c].dataType, StringType) else col(c)
        for c in df.columns
    ])

crm_cust_info = trim_all_columns(crm_cust_info)
erp_cust_info = trim_all_columns(erp_cust_info)
erp_cust_loc  = trim_all_columns(erp_cust_loc)

# --------------------------------------------------
# 4. STANDARDIZE KEYS (JOIN CONSISTENCY)
# --------------------------------------------------
erp_cust_info = erp_cust_info.withColumn(
    "cst_key", regexp_replace(col("CID"), "NAS", "")
)

erp_cust_loc = erp_cust_loc.withColumn(
    "cst_key", regexp_replace(col("CID"), "-", "")
)

# --------------------------------------------------
# 5. JOIN DATASETS (USE BROADCAST FOR SMALL TABLES)
# --------------------------------------------------
df = crm_cust_info.join(
    broadcast(erp_cust_info.select("cst_key","BDATE","GEN")),
    on="cst_key",
    how="left"
)

df = df.join(
    broadcast(erp_cust_loc.select("cst_key","CNTRY")),
    on="cst_key",
    how="left"
)

# --------------------------------------------------
# 6. REPARTITION FOR PARALLEL PROCESSING
# --------------------------------------------------
df = df.repartition(8, "cst_key")  # adjust partitions based on cluster size

# --------------------------------------------------
# 7. CACHE (AVOID RECOMPUTATION)
# --------------------------------------------------
df = df.cache()
df.count()  # trigger cache

# --------------------------------------------------
# 8. DATA TYPE CONVERSIONS
# --------------------------------------------------
df = df.withColumn("cst_create_date", to_date(col("cst_create_date")))
df = df.withColumn("BDATE", to_date(col("BDATE")))

# --------------------------------------------------
# 9. TRANSFORMATIONS (COMBINED FOR PERFORMANCE)
# --------------------------------------------------
df = df.withColumn(
    "cst_marital_status",
    when(col("cst_marital_status") == "M", "Married")
    .when(col("cst_marital_status") == "S", "Single")
    .otherwise("Unknown")
)

df = df.withColumn(
    "gender",
    when(upper(col("GEN")) == "MALE", "Male")
    .when(upper(col("GEN")) == "FEMALE", "Female")
    .when(upper(col("cst_gndr")) == "M", "Male")
    .when(upper(col("cst_gndr")) == "F", "Female")
    .otherwise("Unknown")
)

# --------------------------------------------------
# 10. CLEAN COUNTRY VALUES
# --------------------------------------------------
df = df.withColumn("CNTRY", trim(col("CNTRY")))

df = df.withColumn(
    "CNTRY",
    when(col("CNTRY").isin("US","USA"), "United States")
    .when(col("CNTRY") == "DE", "Germany")
    .when((col("CNTRY") == "") | col("CNTRY").isNull(), "Unknown")
    .otherwise(col("CNTRY"))
)

# --------------------------------------------------
# 11. FILTER INVALID RECORDS EARLY
# --------------------------------------------------
df = df.filter(
    col("cst_id").isNotNull() & col("cst_key").isNotNull()
)

# --------------------------------------------------
# 12. CREATE CUSTOMER NAME
# --------------------------------------------------
df = df.withColumn(
    "cust_name",
    concat_ws(" ", col("cst_firstname"), col("cst_lastname"))
)

df = df.withColumn(
    "cust_name",
    when((col("cust_name") == "") | col("cust_name").isNull(), "Unknown")
    .otherwise(col("cust_name"))
)

# --------------------------------------------------
# 13. SELECT FINAL COLUMNS
# --------------------------------------------------
final_df = df.select(
    "cst_id",
    "cst_key",
    "cust_name",
    "cst_marital_status",
    "gender",
    "BDATE",
    "CNTRY",
    "cst_create_date"
)

# --------------------------------------------------
# 14. REMOVE DUPLICATES
# --------------------------------------------------
final_df = final_df.dropDuplicates(["cst_id"])

# --------------------------------------------------
# 15. DATA QUALITY CHECKS
# --------------------------------------------------
from pyspark.sql.functions import sum as spark_sum

# Null counts
final_df.select([
    spark_sum(col(c).isNull().cast("int")).alias(c)
    for c in final_df.columns
]).show()

# Duplicate check
print("Duplicates:", final_df.count() - final_df.dropDuplicates(["cst_id"]).count())

# Total records
print("Total records:", final_df.count())

# Preview
final_df.show(5)

# --------------------------------------------------
# 16. WRITE OUTPUT (OPTIMIZED STORAGE)
# --------------------------------------------------
final_df.write \
    .mode("overwrite") \
    .partitionBy("CNTRY") \
    .parquet("/content/clean_customer_data")

# Optional: control file size
# final_df.coalesce(4).write.mode("overwrite").parquet("/content/output")
```
